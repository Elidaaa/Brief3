```python
# Faire un rapprochement entre l’export de la table de CMS et l’export de l’ERP

import pandas as pd
```

```python
#Créer DataFrame_erp

df_erp = pd.read_excel('erp.xlsx')
df_erp
```

```python
#Vérifier s'il y a des colonne vides

df_erp.info()
```

```python
#Créer DataFrame_liaison

df_liaison = pd.read_excel('liaison.xlsx')
df_liaison
```

```python
#Vérifier s'il y des colonnes sans valeur

df_liaison.info()
```

```python
# Il y a des colonnes sans id_web
```

```python
#Enlever les colonnes sans id_web

df_liaison = df_liaison[df_liaison["id_web"].notnull()]
df_liaison
```

```python
#Vérification

df_liaison.info()
```

```python
#Il n'y a plus de colonnes sans id_web
```

```python
#Lier DataFrame_erp avec DataFrame_Liaison : merged_df

merged_df = df_erp.merge(df_liaison)
merged_df
```

```python
merged_df.info()
```

```python
#Création DataFrame_web

df_web = pd.read_excel('web.xlsx')
df_web
```

```python
#Renomer "sku" du DataFrame_web en "id_web" afin de pouvoir lier DataFrame_web au nouveau DataFrame (merged_df) 

df_web.rename(columns = {'sku':'id_web'}, inplace = True)
```

```python
#Vérifier que la modification a bien eu lieu

print(df_web.columns)
```

```python
#Nous avons bien 'id_web' à la place de 'sku'.
```

```python
#Vérifier s'il y a des lignes sans id_web

df_web.info()
```

```python
#On peut voir qu'il y a des lignes sans valeur dans la colonne 'id_web' : 1428 non-null.
#Enlever ces lignes.

df_web = df_web[df_web["id_web"].notnull()]
df_web
```

```python
#Vérification

df_web.info()
```

```python
#Lier merged_df avec df_web : table_final_df

table_final_df = merged_df.merge(df_web)
table_final_df
```

```python
#Dans la colonne 'post_type' on garde que les lignes 'product', car les lignes 'attachment' sont des doublons qui contienne juste des images en plus
#donc qui ne sont pas necessaire

table_final_df['post_type'].value_counts()
```
